## Exploratory Data Analysis
This is to veryfy what are the most common words used in the text reviews using a word cloud. However, since the preprocessing has not been doen before, one might view some words that give little meaning to the sentiment of the text.

```python
rt = df_amazon_alexa['verified_reviews']
wordcloud = WordCloud(background_color='white',
                      width=1100,
                      height=500
                     ).generate(" ".join(rt))
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.title('All Words in the Reviews\n',size=20)
plt.axis('off')
plt.show()
```

Of course, as expected there are some words that have little or no importance with the sentiment of the comment, such as: use, device, time etc. Thus, there is need  to perform a word count of certain key words that could describe the sentiment in a better way.

```python
words = ['awesome','great','fantastic','extraordinary','amazing','super',
                 'magnificent','stunning','impressive','wonderful','breathtaking',
                 'love','content','pleased','happy','glad','satisfied','lucky',
                 'shocking','cheerful','wow','sad','unhappy','horrible','regret',
                 'bad','terrible','annoyed','disappointed','upset','awful','hate']

rt = " ".join(df_amazon_alexa['verified_reviews'])
```

```python
word_count = {}
for word in rt.split(" "):
    if word in words:
        word_count[word] = word_count.get(word, 0)+1
```

```python
wordcloud = WordCloud(background_color='white',
                      width=1100,
                      height=500
                     ).generate_from_frequencies(word_count)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.title('Sentiment Words\n',size=20)
plt.axis('off')
plt.show()
```

```python
ps = PorterStemmer()
reviews=[]
for i in range(0,3150):
    review = re.sub('[^a-zA-Z]',' ',df_amazon_alexa['verified_reviews'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    reviews.append(review)
```

```python
#Creating a bag of words that contains all the unique words in all the reviews
#Each column in X will represent each word
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(reviews).toarray()
y = df_amazon_alexa.iloc[:,4]
```

```python
# Slit the data into train and test set
from sklearn.model_selection import train_test_split
xtr, xtst, ytr, ytst = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# Building a Naive bayes classifier for classification 
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()

# fitting the train data into the classifier
classifier.fit(xtr, ytr)
classifier.score(xtr, ytr)
```

```python
#Testing the classifier onto test data and storing the results in y_pred variable
y_pred = classifier.predict(xtst)
```

```python
#Checking accuracy of predictions on test data
#Using confusion matrix from sklearn.metrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytst,y_pred)
print((cm[0,0]+cm[1,1])/cm.sum()*100)
```

It is apparent the prediction is not very good enough as it's depicted above. Calculating the correct % accuracy it will be 55% only. Thus, the need to use another model which is Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
classifier_dt = DecisionTreeClassifier(criterion='entropy', random_state=0)

# Classifier training 
classifier_dt.fit(xtr,ytr)
classifier_dt.score(xtr, ytr)
```

```python
# Test data prediction 
y_pred_dt = classifier_dt.predict(xtst)
```

```python
# Check prediction accuracy
prediction_accuracy = confusion_matrix(ytst,y_pred_dt)
print((prediction_accuracy[0,0]+prediction_accuracy[1,1])/prediction_accuracy.sum()*100)
```

## Accuracy prediction 
Eveidently. the confusion matrix clearly show that the built model is indeed a great one for the data, the model is 92% accurate if you will calculate the score. So Decision Tree classifier predictions are far better than Naive Bayes Classifier.
